```python
import pandas as pd, os, numpy as np
```

```python
def loadData(folder):
    frames = {}
    for year_label in os.listdir(folder):
        if not os.path.isdir(folder + year_label):
            continue
        for excel in os.listdir(folder + year_label):
            print "\nimporting " + folder + year_label + "/" + excel
            frames[year_label] = pd.read_excel(folder + year_label + "/" + excel)
            frames[year_label].columns = map(lambda x: x.strip().upper(), frames[year_label].columns)
    return frames

fiscal_frames = loadData("data/Fiscal Links.html/")
nonfiscal_frames = loadData("data/Nonfiscal links.html/")
```

```python
def get_column(col_name, frame_list):
    cols = "STABR"
    ret_frame = pd.DataFrame({cols: []})
    for df in frame_list:
        if col_name not in frame_list[df].columns:
            print "** Skipping " + df + " **"
            continue
        if "STABR" not in frame_list[df].columns:
            print "** Skipping " + df + " **"
            continue
        #print frame_list[df]
        # ret_frames[df] = frame_list[df][["STABR",col_name]]
        cur_frame = frame_list[df][["STABR",col_name]]
        cur_frame.columns = cols, df
        #cur_frame  = cur_frame[cur_frame[df].str.isnumeric().isnull()]
        #cur_frame = cur_frame[cur_frame[df].str.isnumeric()]
        #cur_frame = cur_frame[cur_frame[df]].applymap(np.isreal)
        cur_frame = cur_frame.set_index("STABR")
        #cur_frame = cur_frame[cur_frame[df] !="R"]
        cur_frame[df] = pd.to_numeric(cur_frame[df], errors="coerce")
        cur_frame = cur_frame[cur_frame[df].notnull()]
        if ret_frame.empty:
            print df, "None"
            #return cur_frame
            ret_frame = cur_frame
        else:
            print "Not none"
            #print ret_frame.head(2)
            #print cur_frame.head(2)
            #return [cur_frame, ret_frame]
            ret_frame = cur_frame.join(ret_frame)

    return ret_frame
```
